```python
import pandas as pd
import numpy as np
import re 
```

```python
df = pd.read_excel('AcademicPlacement.xlsx')
```

```python
df['Source'] = df['Source'].str.lower()
df['Placement'] = df['Placement'].str.lower()
df['Scores'] = df['Score']*df['Discount']
```

```python
uni_list = df['Source'].unique()
```

```python
len(uni_list)
```

```python
df.shape
```

```python
def clean_target(df):
    '''
    The function cleans the Placement column by keeping the employers' names
    '''
    item = df['Placement']
    
    item = item.replace('assistant professor', '')
    item = item.replace('lecturer', '')
    

    if ' – ' in item:
        item = item.replace(' – ', ',')
    if '(' in item:
        item = item.replace('(', ',')
    if ')' in item:
        item = item.replace(')', ',')     
    if ',' in item:
        item = item.split(',')
 
        
    
    #print(type(item)==str)
    if type(item)==str:
        
        return item
    
    else:
        for i in item:
            if 'university' in i or 'institute' in i \
                or 'state' in i or 'college' in i \
                or 'school' in i or 'kellogg ' in i \
                or 'universidad' in i or 'polytechnique' in i \
                or 'instituto ' in i or 'fgv rio' in i:

                return i 

        

        
```

```python
df['employers'] = df.apply(clean_target, axis=1)
```

```python
df.sample(10)
```

```python
glo = df[['ID', 'Source', 'Year', 'employers', 'Scores']]
glo.rename(columns={'employers':'Target'}, inplace=True)
glo.to_csv('GlobalPlacement.csv')
```

### Top 100 US

```python
def usnodes(df):
    '''
    The function returns indicators of whether the employers if it is one of the 104 major U.S schools
    '''
    for i in uni_list:
        try:
            if df['employers'] in i or i in df['employers']:
                return 1

        except TypeError:
            return 0
```

```python
def usemployers(df):
    '''
    The function keeps the employers if it is one of the 104 major U.S schools
    '''
    if df['USnodes']==1:
        for i in uni_list:
            try:
                if df['employers'] in i or i in df['employers']:
                    return i
            except TypeError:
                return 0
    else:
        return 0
        
```

```python
df['USnodes'] = df.apply(usnodes, axis=1)
df['USnodes'] = df['USnodes'] .fillna(0)
```

```python
df.sample(10)
```

```python
df['Target'] = df.apply(usemployers, axis=1)
```

```python
agg = df.groupby(['Source','Year']).agg({'USnodes':'mean', 
                                'Scores':'mean',
                                'ID': 'count'})
```
